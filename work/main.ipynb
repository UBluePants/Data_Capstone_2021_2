DIY dataloader

```python
import torch.utils.data as torch_data
from torchvision import transforms
from data.YJ_dataloader import YJ_Dataloader, loader
from options.base_options import BaseOptions


opt = BaseOptions().parse()

train_transform = transforms.Compose([
            transforms.Resize(128),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.5, 0.5, 0.5],
                                std=[0.5, 0.5, 0.5])
        ])
test_transform = transforms.Compose([
            transforms.Resize(128),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.5, 0.5, 0.5],
                                std=[0.5, 0.5, 0.5])
        ])



dataset_train = torch_data.DataLoader(dataset=YJ_Dataloader(
                                            opt.GT_train_root,
                                            opt.Synthetic_train_root,
                                            opt.non_occ_train_root,
                                            opt.occ_train_root,                       
                                            transform = train_transform),
                                            batch_size= opt.batch_size, 
                                            num_workers = 0,
                                            shuffle=True, drop_last=True)

dataset_test = torch_data.DataLoader(dataset=YJ_Dataloader(
                                            opt.GT_test_root, 
                                            opt.Synthetic_test_root,
                                            opt.non_occ_test_root,
                                            opt.occ_test_root, 
                                            transform=test_transform),
                                            batch_size=opt.batch_size,
                                            drop_last=True)
                                            


```
