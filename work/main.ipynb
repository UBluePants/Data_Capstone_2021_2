```python
from models.modules.generator_wgan import Generator
import os
import IQA_pytorch
import torch

model = Generator()

load_filename = 'net_epoch_%s_id_%s.pth' % (50, 'G')
load_path = os.path.join("/content/drive/Othercomputers/내 노트북/data/check_points/Experiment_1125", load_filename)
model.load_state_dict(torch.load(load_path))
model = model.cuda()

tf = transforms.ToPILImage()
D = IQA_pytorch.SSIM()

def PSNR(GT, Target):
  result = 20 * torch.log10(255 / torch.sqrt(torch.mean((Target - GT) ** 2)))
  return result.item()

def SSIM(GT, Target):
  score = 0
  GT_img = tf(GT)
  rec_img = tf(Target)            
  GT_img = IQA_pytorch.utils.prepare_image(GT_img).cuda()
  rec_img = IQA_pytorch.utils.prepare_image(rec_img).cuda()
  score = D(GT_img, rec_img, as_loss = False).item()
  return score
  
```

```python
PSNR_score = 0 
SSIM_score = 0 

for i, j in enumerate(dataset_test):
  GT = j["GT_img"]
  occ_img = j["synth_img"]
  GT, occ_img = GT.cuda(), occ_img.cuda()
  fake_img, fake_img_mask = model.forward(occ_img)
  fake_img_synthesis = fake_img_mask * occ_img + (1 - fake_img_mask) * fake_img
  PSNR_batch = PSNR(GT, fake_img_synthesis)
  SSIM_batch = SSIM(GT, fake_img_synthesis)
  print("batch ", i, " PSNR : " , PSNR_batch , " SSIM : ", SSIM_batch)
  PSNR_score = PSNR_score + PSNR_batch
  SSIM_score = SSIM_score + SSIM_batch

print("Mean of 50 epoch generator : PSNR = ", PSNR_score / i+1, " SSIM = ", SSIM_score / i+1)
```

```python
from PIL import Image
import matplotlib.pyplot as plt 
from utils.util import tensor2im
psnr_tot = 0
ssim_tot = 0
dif_psnr_tot = 0
dif_ssim_tot = 0
for i, j in enumerate(dataset_test):
  GT = j["GT_img"]
  occ_img = j["synth_img"]
  
  GT, occ_img = GT.cuda(), occ_img.cuda()
  fake_img, fake_img_mask = model.forward(occ_img)
  fake_img_synthesis = fake_img_mask * occ_img + (1 - fake_img_mask) * fake_img
  GT_img = tensor2im(GT, idx=-1)
  rec_img = tensor2im(fake_img_synthesis, idx=-1)
  occ_img = tensor2im(occ_img, idx=-1)
  psnr_before = PSNR(torch.tensor(GT_img, dtype=torch.float64), torch.tensor(occ_img, dtype=torch.float64))
  ssim_before = SSIM(GT_img.transpose(1,2,0), occ_img.transpose(1,2,0))
  psnr_after = PSNR(torch.tensor(GT_img, dtype=torch.float64), torch.tensor(rec_img, dtype=torch.float64))
  ssim_after = SSIM(GT_img.transpose(1,2,0), rec_img.transpose(1,2,0))
  diff_psnr = psnr_after - psnr_before
  diff_ssim = ssim_after - ssim_before
  print("batch ", i +1)
  print("PSNR : ", psnr_before, "-->", psnr_after, "difference : ", diff_psnr)
  print("SSIM : ", ssim_before, "-->", ssim_after, "difference : ", diff_ssim)
  psnr_tot += psnr_after
  ssim_tot += ssim_after
  dif_psnr_tot += diff_psnr
  dif_ssim_tot += diff_ssim

print("final | PSNR : ", psnr_tot / (i + 1), "SSIM : ", ssim_tot / (i + 1))
print("mean of difference PSNR : ",dif_psnr_tot / (i +1))
print("mean of difference SSIM : ",dif_ssim_tot / (i +1))
```
